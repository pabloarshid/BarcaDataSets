```python
%matplotlib inline
import os
import re
import pandas as pd
import datetime as dt
import numpy as np
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [16, 10]
plt.rcParams['font.size'] = 14
import seaborn as sns
import warnings
fig = plt.gcf()
fig.set_size_inches( 16, 10)
```

```python
busstops= pd.read_csv("bus_stops.csv")
transports = pd.read_csv("transports.csv")
accidents = pd.read_csv("accidents_2017.csv")
population = pd.read_csv("population.csv")
births = pd.read_csv("births.csv")
deaths = pd.read_csv("deaths.csv")
airq= pd.read_csv("air_quality_Nov2017.csv")
airstat= pd.read_csv("air_stations_Nov2017.csv")
immbynat = pd.read_csv("immigrants_by_nationality.csv")
immbyage = pd.read_csv("immigrants_emigrants_by_age.csv")
immbyd1 = pd.read_csv("immigrants_emigrants_by_destination.csv")
immbyd2 = pd.read_csv("immigrants_emigrants_by_destination2.csv")
immbysex = pd.read_csv("immigrants_emigrants_by_sex.csv")

freqnames = pd.read_csv("most_frequent_names.csv")
freqbabynames = pd.read_csv("most_frequent_baby_names.csv")
unemployement = pd.read_csv("unemployment.csv")
```

```python
populationbyyear = population.groupby(["Year","Gender"], as_index=False).sum()
populationbyD = population.groupby(["Year","District.Name"], as_index=False).sum()
```

```python
ax = sns.barplot(x = "Year", y = "Number", hue = "Gender", data = populationbyyear)
ax.set(ylim=(700000, 900000))
```

```python
ax = sns.barplot(x = "Year", y = "Number", hue = "District.Name", data = populationbyD)
ax.set(ylim=(50000, 300000))
```

```python
accbymonthcts = accidents.groupby('Weekday').Hour.value_counts()
accbymonthcts.to_frame().Hour
accbymonthcts = accbymonthcts.reset_index(level=["Weekday"])
accbymonthcts = accbymonthcts.rename(columns = {'Hour':'Count'})
accbymonthcts.reset_index(level=0, inplace=True)

sns.catplot("Weekday","Count", "Hour", data = accbymonthcts, kind = "bar", palette="muted",height=8.27, aspect=11.7/8.27)
```

```python
accbydistcts = accidents.groupby('District Name').Hour.value_counts()
accbydistcts.to_frame().Hour
accbydistcts = accbydistcts.reset_index(level=["District Name"])
accbydistcts = accbydistcts.rename(columns = {'Hour':'Count'})
accbydistcts.reset_index(level=0, inplace=True)

sns.catplot("District Name","Count", "Hour", data = accbydistcts, kind = "bar", palette="muted",height=8.27, aspect=20.7/8.27)
```

```python
accEix = accidents[accidents["District Name"] == "Eixample"]
accEix["Neighborhood Name"].unique()

accinEix = accEix.groupby('Neighborhood Name').Hour.value_counts()
accinEix.to_frame().Hour
accinEix = accinEix.reset_index(level=["Neighborhood Name"])
accinEix = accinEix.rename(columns = {'Hour':'Count'})
accinEix.reset_index(level=0, inplace=True)


sns.catplot("Neighborhood Name","Count", "Hour", data = accinEix, kind = "bar", palette="muted",height=8.27, aspect=20.7/8.27)
```
